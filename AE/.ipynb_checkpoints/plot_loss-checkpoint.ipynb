### Loss value changes with network depth

```python
import csv
import matplotlib.pyplot as plt

filename_list = ['./epoch_train_test_param/AE_MSE_ReLU_0.01_4.csv',
                 './epoch_train_test_param/AE_MSE_ReLU_0.01_80_4.csv',
                 './epoch_train_test_param/AE_MSE_ReLU_0.01_220_30_4.csv',
                 './epoch_train_test_param/AE_MSE_ReLU_0.01_360_80_18_4.csv',
                 './epoch_train_test_param/AE_MSE_ReLU_0.01_480_140_40_12_4.csv',
                 './epoch_train_test_param/AE_MSE_ReLU_0.01_580_210_80_30_12_4.csv',
                 './epoch_train_test_param/AE_MSE_ReLU_0.01_680_280_120_50_22_9_4.csv',
                 './epoch_train_test_param/AE_MSE_ReLU_0.01_790_360_170_80_38_18_8_4.csv',
                 './epoch_train_test_param/AE_MSE_ReLU_0.01_830_430_220_110_60_30_15_8_4.csv']

x_data = range(2,11)
y_train = []
y_test = []

for filename in filename_list:
    with open(filename, 'r') as file:
        reader = csv.reader(file)
        next(reader)

        train_loss = 0
        test_loss = 0
        epoch_n = 0
        for row in reader:
            train_loss += float(row[1])
            test_loss += float(row[2])
            epoch_n = float(row[0])

        y_train.append(train_loss/epoch_n)  
        y_test.append(test_loss/epoch_n)  
   
plt.plot(x_data, y_train, label='train', marker='P', linewidth=2)
plt.plot(x_data, y_test, label='test', marker='D', linestyle='--', linewidth=2)
plt.xlabel('depth of AE')
plt.ylabel('Average MSE loss')
plt.title('Average MSE loss of AE with different number of layers')
plt.legend()
plt.savefig('./pic_loss/Average_Loss_depth.png', dpi=300)
plt.show()
```

### Loss value changes with learning rate

```python
import csv
import matplotlib.pyplot as plt

filename_list = ['./epoch_train_test_param/AE_MSE_ReLU_0.02_80_4.csv',
                 './epoch_train_test_param/AE_MSE_ReLU_0.01_80_4.csv',
                 './epoch_train_test_param/AE_MSE_ReLU_0.005_80_4.csv',
                 './epoch_train_test_param/AE_MSE_ReLU_0.0025_80_4.csv']

i = 0
lr = 0.02
for filename in filename_list:
    x_data = []
    y_data = []

    with open(filename, 'r') as file:
        reader = csv.reader(file)
        next(reader)

        for row in reader:
            x_data.append(float(row[0]))  # 假设X轴数据在CSV文件的第一列
            y_data.append(float(row[1]))  # 假设Y轴数据在CSV文件的第二列
    
    plt.plot(x_data, y_data, label=str(lr), color=plt.cm.tab10(i), marker='P', linewidth=2)
    i += 1
    lr /= 2

plt.xlabel('epoch')
plt.ylabel('MSE loss')
plt.title('MSE train loss in every epoch \n during training with different learning rates')
plt.legend()
plt.savefig('./pic_loss/MSE_train_loss_epoch_LR.png', dpi=300)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
import numpy as np

filename_list = ['./epoch_train_test_param/AE_MSE_ReLU_0.02_80_4.csv',
                 './epoch_train_test_param/AE_MSE_ReLU_0.01_80_4.csv',
                 './epoch_train_test_param/AE_MSE_ReLU_0.005_80_4.csv',
                 './epoch_train_test_param/AE_MSE_ReLU_0.0025_80_4.csv']

i = 0
lr = 0.02
for filename in filename_list:
    x_data = []
    y_data = []

    with open(filename, 'r') as file:
        reader = csv.reader(file)
        next(reader)

        for row in reader:
            x_data.append(float(row[0]))  # 假设X轴数据在CSV文件的第一列
            y_data.append(float(row[2]))  # 假设Y轴数据在CSV文件的第二列
    
    plt.plot(x_data, y_data, label=str(lr), color=plt.cm.tab10(i), marker='P', linewidth=2)
    i += 1
    lr /= 2

plt.xlabel('epoch')
plt.ylabel('MSE loss')
plt.yticks(np.arange(0.01,0.05,0.005))
plt.title('MSE test loss in every epoch \n during training with different learning rates')
plt.legend()
plt.savefig('./pic_loss/MSE_test_loss_epoch_LR.png', dpi=300)
plt.show()
```

### Loss value changes with the number of neurons in the hidden space

```python
import csv
import matplotlib.pyplot as plt

y_train = []
y_test = []
x_data = range(2,72,2)
for NLS in range(2,72,2):
    filename = "./epoch_train_test_param/AE_MSE_ReLU_0.01_"+str(NLS)+".csv"
    with open(filename, 'r') as file:
        reader = csv.reader(file)
        next(reader)

        test_err = 0
        train_err = 0
        epoch_n = 0
        for row in reader:
            train_err += float(row[1])  # 假设Y轴数据在CSV文件的第二列
            test_err += float(row[2])  # 假设Y轴数据在CSV文件的第二列
            epoch_n += 1
        y_train.append(train_err/epoch_n)
        y_test.append(test_err/epoch_n)
        # print(train_err/epoch_n, test_err/epoch_n)

plt.plot(x_data, y_train, label="train", linewidth=2)
plt.plot(x_data, y_test, label="test", linestyle="--", linewidth=2)
plt.xlabel('Number of neurons in latent space')
plt.ylabel('Average MSE loss')
plt.title('Average loss of autoencoders with \n different numbers of neurons in latent space')
plt.legend()
plt.savefig('./pic_loss/Average_loss_NLS_2-70.png', dpi=300)
plt.show()
```
